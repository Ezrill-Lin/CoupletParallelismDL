```python
import os
import pandas as pd
```

```python
# Define the directory containing the CSV files
dir_path = r'C:\Users\linxi\OneDrive\Desktop\NYU\Computational Cognitive Modelling\project\分词数据'

# Create a list to store dfs
dfs = []

# Iterate through the files in the directory
for filename in os.listdir(dir_path):
    file_path = os.path.join(dir_path, filename)
    try:
        # Read the CSV file into a df
        df = pd.read_csv(file_path)
        dfs.append(df)  # Add the df to the list
    except Exception as e:
        print(f"Error reading {filename}: {e}")

# Combine all dfs into one 
combined_df = pd.concat(dfs, ignore_index=True)

# Print the combined df (or save it to a file)
print(combined_df)
```

```python
# These matched data are used to train the word segmentation model
matched_data = combined_df[combined_df['Column1_Tag']==combined_df['Column2_Tag']]
matched_data.to_csv('matched_data.csv')
matched_data
```

```python
unmatched_data = combined_df[combined_df['Column1_Tag']!=combined_df['Column2_Tag']]
unmatched_data.to_csv('unmatched_data.csv')
unmatched_data
```
